```python
Days = ["sun","mon","tues","wednes","thurs","fri","stur"] #Days of the Week
counter = 0                                               #Counter days
for year in range (1400,1401):                            #Range year
    for month in range (1,7):                             #Range the first half of the month 
        for day in range (1,32):                          #Range days 
            print(year,"/",month,"/",day,"  ", Days[counter])
            counter += 1 
            while counter > 6:
                counter = 0                               #Start counter 
    for month in range (7,13):                            #Range the second half of the month
        for day1 in range (1,31):                         #Range days
            print(year,"/",month,"/",day1,"  ",Days[counter])
            counter += 1 
            while counter > 6:
                counter = 0                              #Start counter 
```
